```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.figure_factory as ff
```

# Autos MPG Dataset

Loads the car dataset from the csv file and store it to a dataframe.

```python
df = pd.read_csv("auto-mpg.csv", header=None, names=["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model year", "origin", "car name"])
df
```

## Box plot using pandas

```python
# Draw the box plot using pandas
boxplot = df.boxplot(column=['mpg'], by=['displacement'], grid=False, fontsize=10, figsize=(20,20), rot=90)
boxplot = df.boxplot(column=['mpg'], by=['cylinders'], grid=False, fontsize=10, figsize=(20,20), rot=90)
boxplot = df.boxplot(column=['mpg'], by=['horsepower'], grid=False, fontsize=10, figsize=(20,20), rot=90)
boxplot = df.boxplot(column=['mpg'], by=['acceleration'], grid=False, fontsize=10, figsize=(20,20), rot=90)
```

## Correlation heatmaps using Seaborn

```python
plt.figure(figsize=(10, 10))
heatmap = sns.heatmap(df.corr(method='kendall'), annot=True)
heatmap.set_title('Kendall Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
plt.show()
```
